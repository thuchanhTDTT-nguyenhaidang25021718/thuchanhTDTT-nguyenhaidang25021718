```python
#bai 9
a = int(input())
b = int(input())
c = int(input())

edge = [a, b, c]
sum = a+b+c
isTriangle = True

for i in edge:
    if i>=sum-i:
        isTriangle = False
        break

if isTriangle:
    print("Yes")
else:
    print("No")
```

```python
#bai 12
def bac(so_dien):
    if 0<=so_dien<=50:
        return 1500
    if 51<=so_dien<=100:
        return 2000
    if so_dien>100:
        return 3000

so_dien = int(input())
tien = 0
while (so_dien!=0):
    tien += bac(so_dien)
    so_dien -= 1

print(tien)
```
