```python
#W6A10
cnt = {}

def counter(x:int, cnt:dict) :
    x = int(x)
    if x in cnt:
        cnt[x]+=1
    else:
        cnt[x]=1
    return x

a = [counter(x, cnt) for x in input().split()]
k = int(input())
ans = []
for i in a:
    if cnt[i]!=0 and k-i in cnt:
        if i==k-i and cnt[i]==1 and cnt[k-i]==1:
            continue
        elif cnt[i]>0 and cnt[k-i]>0:
            cnt[i]-=1
            cnt[k-i]-=1
            ans.append((i, k-i))

ans.sort()
print(ans)
```

```python
#W6A19
def find_max(a:list[int])->int:
    idx = 0
    maxval = a[0]
    for i in range(1, len(a)):
        if maxval<a[i]:
            maxval = a[i]
            idx = i
    return idx+1

a = [int(x) for x in input().split()]

print(find_max(a))
```

```python
#W6A20
def find_k(a:list[int], k):
    for idx, i in enumerate(a):
        if i==k:
            return idx+1
    return -1

a = [int(i) for i in input().split()]
k = int(input())

print(find_k(a, k))
```
