```python
#W7A1
def binary_search(arr, left, right, target)->int:
    a = arr
    l = left
    r = right
    b = target
    while l<=r:
        mid = l+ (r-l)//2
        if a[mid]==b:
            return mid
        elif a[mid]<b:
            l = mid + 1
        else:
            r = mid - 1
    return -1

a = [int(x) for x in input().split()]
b = int(input())
print(binary_search(a, 0, len(a)-1, b))
```

```python
#W7A3
def bubble_sort(arr):
    for i in range (len(arr)-1):
        swap = False
        for j in range(len(arr)-i-1):
            if arr[j]>arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
                swap = True
        if not swap: return

def selection_sort(arr):
    for i in range(len(arr)):
        minval = arr[i]
        minidx = i
        for j in range(i, len(arr)):
            if arr[j]<minval:
                minval = arr[j]
                minidx = j
        arr[i], arr[minidx] = arr[minidx], arr[i]

a = [int(x) for x in input().split()]
a1 = a.copy()
a2 = a.copy()
bubble_sort(a1)
print(a1)
selection_sort(a2)
print(a2)
```

```python
#W7A4
def count_max(arr):
    d = {}
    for x in arr:
        if x in d:
            d[x]+=1
        else:
            d[x]=1
    maxnum = 0
    maxcnt = -999999999
    for x in d:
        if d[x]>maxcnt:
            maxnum = x
            maxcnt = d[x]
    return maxnum, maxcnt

a = [int(x) for x in input().split()]
maxnum, maxcnt = count_max(a)
print(maxnum, "xuat hien nhieu nhat, som nhat,", maxcnt, "lan")
```

```python
#W7A5
d = {}

a = [int(x) for x in input().split()]
c = int(input())

for i in range(len(a)):
    if a[i] in d:
        d[a[i]]+=1
    else:
        d[a[i]]=1

ans = 0
calculated_pair = {}
for i in a:
    if c - i in a:
        if (i, c-i) in calculated_pair or (c-i, i) in calculated_pair:
            continue
        if i==c-i:
            cnt = d[i]
            ans += cnt*(cnt-1)//2
        else:
            cnt = d[i]
            cnt2 = d[c-i]
            ans += cnt*cnt2
        calculated_pair[(i, c-i)] = True
        calculated_pair[(c-i, i)] = True

print(ans)
```

```python
#W7A8
def bs(arr, target):
    #lower bound
    l = 0
    r = len(arr)
    while l<r:
        mid = (l+r)//2
        if arr[mid][0]>=target:
            r = mid
        else:
            l = mid+1
    if l<len(arr):
        return l
    return -1

interval = eval(input())
queries = eval(input())

w = []
for i in interval:
    w.append((i[1]-i[0]+1, i[0], i[1]))
w.sort()

qsort = []
negq = []
for idx, i in enumerate(queries):
    qsort.append((i, idx))
    negq.append((-i, idx))
qsort.sort()
negq.sort()

#print(qsort)

ans = [-1]*len(queries)
for idx, i in enumerate(w):
    #if any -1 => no sol
    #else from idx1 to idx2 in qsort, qsort[i]=idx
    idx1 = bs(qsort, i[1])
    idx2 = bs(negq, -i[2])
    idx2 = len(queries)-idx2-1
    if idx1==-1 or idx2==-1:
      continue
    for j in range(idx1, idx2+1):
        if ans[qsort[j][1]]==-1:
            ans[qsort[j][1]]=i[0]

print(ans)
```

```python
#W7A9

nums = eval(input())

#nen so
val = sorted(set(nums))
d = {v:i for i,v in enumerate(val)}
for i in range(len(nums)):
    nums[i] = d[nums[i]]

#segtree
n = len(val)
seg = [0] * (2 * n)

def rsq(l, r):
    global seg, n
    l+=n
    r+=n
    ans = 0
    while l<=r:
        if l&1:
            ans+=seg[l]
            l+=1
        if not r&1:
            ans+=seg[r]
            r-=1
        l>>=1
        r>>=1
    return ans

def update(idx, add):
    #add to cnt[idx]
    global seg, n
    idx+=n
    seg[idx]+=add
    while idx>>1:
        seg[idx>>1]+=add
        idx>>=1

#sol
ans = [0] * len(nums)
for i in range(len(nums)-1, -1, -1):
    # sum of all cnt in range 0..nums[i]-1
    update(nums[i], 1)
    ans[i] = rsq(0, nums[i]-1)

print(ans)
```

```python
#W7A10

def bs(arr, target):
    #upper
    if not arr:
        return -1
    l, r = 0, len(arr)
    while l<r:
        m = (l+r)//2
        if arr[m]>target:
            r = m
        else:
            l = m+1
    if l<len(arr):
        return l
    return len(arr)

nums = eval(input())
m = int(input())

val = [0]*m
prefix = [0]*(len(nums)+1)

for i in range(1, len(nums)+1):
    prefix[i] = (nums[i-1]%m+prefix[i-1])%m

#rsq i..j: (prefix[j] - prefix[i-1] + m)%m
#voi moi mot vi tri j ta can tim vi tri i thoa man rsq max

t = []
ans = -999999999

for i in range(len(prefix)):
    best = bs(t, prefix[i])

    if best==-1 or best==len(t):
        ans = max(ans, prefix[i])
        t.append(prefix[i])
    else:
        ans = max(ans, (prefix[i]-t[best]+m)%m)
        t.insert(best, prefix[i])

print(ans)
```

```python
#W7A11
def prefix(s:str)->list[int]:
    pi=[0]*len(s)
    for i in range(1, len(s)):
        j = pi[i-1]
        while j>0 and s[i]!=s[j]:
            j = pi[j-1]
        if s[i]==s[j]:
            j+=1
        pi[i] = j
    return pi

text = input()
key = input()
exist = [False]*128
for i in text:
    exist[ord(i)] = True
for i in key:
    exist[ord(i)] = True
sepa = '#'
for i in range(128):
    if not exist[i]:
        sepa = chr(i)
        break

pi = prefix(key+sepa+text)
ans = 0
for i in range(len(key)+1, len(pi)):
    if pi[i] == len(key):
        ans+=1

print(ans)
```

```python
#W7A13
def merge_sort(a):
    n= len(a)
    if n<2:
        return a
    t = [0]*n
    w = 1
    while w<n:
        for l in range(0, n, w<<1):
            m = min(n, l+w)
            r = min(n, l+(w<<1))
            i,j,k = l,m,l
            while i<m and j<r:
                if a[i]<a[j]:
                    t[k] = a[i]
                    i+=1
                else:
                    t[k] = a[j]
                    j+=1
                k+=1
            while i<m:
                t[k] = a[i]
                i+=1
                k+=1
            while j<r:
                t[k] = a[j]
                j+=1
                k+=1
        a[:] = t[:]
        w<<=1
    return a

a = [int(x) for x in input().split()]
cnt = {}
for i in a:
    if i in cnt:cnt[i]+=1
    else: cnt[i]=1
k = merge_sort(list(cnt.keys()))
ans = 0
for i in range(1, len(k)):
    if k[i]-k[i-1]==1:
        ans = max(ans, cnt[k[i]]+cnt[k[i-1]])

print(ans)
```
