```python
#W8A3
try:
    a = int(input())
    seen = set()

    while a != 1:
        if a in seen:
            print("NO")
            break
        seen.add(a)

        s = 0
        while a > 0:
            digit = a % 10
            s += digit * digit
            a //= 10

        a = s
    else:
        print("YES")

except ValueError:
    print("NO")
```
