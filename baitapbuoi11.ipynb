```python
#W8A3
try:
    a = int(input())
    seen = set()

    while a != 1:
        if a in seen:
            print("NO")
            break
        seen.add(a)

        s = 0
        while a > 0:
            digit = a % 10
            s += digit * digit
            a //= 10

        a = s
    else:
        print("YES")

except ValueError:
    print("NO")
```

```python
#w10a7
s = input()
txt = '.txt'
zip = '.zip'
is_txt = True
is_zip = True
try:
  for i in range(-1, -5, -1):
    if s[i].lower() != txt[i]:
      is_txt = False
      break
  for i in range(-1, -5, -1):
    if s[i].lower() != zip[i]:
      is_zip = False
      break
  if is_txt or is_zip:
    print('Doc file thanh cong')
  else:
    print('File khong hop le')
except IndexError:
    print('File khong hop le')
```

```python
#w10a10
a = [int(x) for x  in input().split()]
exist = {}
valid = True
for i in a:
  if i in exist:
    valid = False
    break
  exist[i] = True
if valid:
  ans = 0
  for i in a:
    ans += i
  print(ans)
else:
  print('Mang khong hop le')
```
