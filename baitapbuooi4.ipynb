```python
#bai 8
def is_palindrome(num) :
    num = str(num)
    length = len(num)
    for i in range (0, length//2):
        if num[i]!=num[length-i-1]:
            return False
    return True

def rev(num) :
    num = str(num)
    length = len(num)
    revnum = ""
    for i in range(0, length):
        revnum=num[i]+revnum
    return int(revnum)

n = int(input())
step=0
while not is_palindrome(n):
    n = n + rev(n)
    step+=1

print(step, n)
```

```python
#bai 10
def collatz(num) :
    count=1
    while(num!=1):
        if num&1==0:
            num//=2
            count+=1
        else :
            num=3*num+1
            count+=1
    return count

maxnum = -999
maxlen = -999
n = int(input())
for i in range (1, n+1) :
    curlen = collatz(i)
    if curlen>maxlen:
        maxnum=i
        maxlen=curlen

print(maxnum, maxlen)
```
