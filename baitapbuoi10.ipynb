```python
#W9A3
def min_after_one_swap(s, i=0, j=1, best=None):
    if i >= len(s) - 1:
        return best

    if j >= len(s):
        return min_after_one_swap(s, i + 1, i + 2, best)

    lst = list(s)
    lst[i], lst[j] = lst[j], lst[i]
    new_num = ''.join(lst)

    if best is None or int(new_num) < int(best):
        best = new_num

    return min_after_one_swap(s, i, j + 1, best)
n = input().strip()
res = min_after_one_swap(n)
print(res.lstrip('0') or '0')
```

```python
#W9A4

def can_convert(a, b):
    memo = {}

    def dfs(i, j):
        if (i, j) in memo:
            return memo[(i, j)]

        if j == len(b):
            for k in range(i, len(a)):
                if a[k].isupper():
                    memo[(i, j)] = False
                    return False
            memo[(i, j)] = True
            return True

        if i == len(a):
            memo[(i, j)] = False
            return False

        if a[i] == b[j]:
            res = dfs(i+1, j+1)
            memo[(i, j)] = res
            return res
        if a[i].islower():
            if dfs(i+1, j):
                memo[(i, j)] = True
                return True
            if a[i].upper() == b[j] and dfs(i+1, j+1):
                memo[(i, j)] = True
                return True

        memo[(i, j)] = False
        return False

    return dfs(0, 0)
Q = int(input())

for _ in range(Q):
    a = input().strip()
    b = input().strip()

    if can_convert(a, b):
        print("YES")
    else:
        print("NO")
```

```python
#W9A6
def quicksort (arr,l,r) :
  mid=int((l+r)/2)
  pivot = arr[mid]
  i,j=l,r
  while(i<j) :
    while(arr[i]<pivot) :
      i+=1
    while(pivot<arr[j]) :
      j-=1
    if i <= j :
      arr[i],arr[j]=arr[j],arr[i]
      i+=1
      j-=1
  if i < r :
    quicksort(arr,i,r)
  if l < j :
    quicksort(arr,l,j)
def hoanvi(a,l,r):
  j=l
  Min=999999999
  for i in range (l+1,r) :
    if (a[l]<a[i]) and (a[i]<Min) :
      j=i
      Min=a[i]
  a[l],a[j]=a[j],a[l]
n=int(input())
dem=0
a=list(map(int,input().split()))
for i in range (n-2,-1,-1) :
  if a[i] < a[i+1] :
    hoanvi(a,i,n)
    quicksort(a,i+1,n-1)
    dem=1
    break
if dem==0 :
  quicksort(a,0,n-1)
  print(a)
else :
  print(a)
```

```python
#W9A10
def findRadius(houses, heaters):
    houses.sort()
    heaters.sort()

    r = 0
    j = 0

    for h in houses:

        while j + 1 < len(heaters) and abs(heaters[j + 1] - h) <= abs(heaters[j] - h):
            j += 1

        r = max(r, abs(heaters[j] - h))

    return r
houses=list(map(int,input().split()))
heaters=list(map(int,input().split()))
findRadius(houses,heaters)
```
