```python
#W8A1
class Cylinder:
    def __init__(self, radius, height):
        self.r = radius
        self.h = height
    def surface_area(self):
        pi = 3.14
        return pi*self.r*self.r + self.h*2*pi*self.r
    def volume(self):
        pi = 3.14
        return pi*self.r*self.r*self.h

radius, height = map(int, input().split())
c = Cylinder(radius, height)

print(f"{c.surface_area():.2f}")
print(f"{c.volume():.2f}")
```

```python
#W8A2
class Date:
    def __init__(self, day, month, year):
        self.d = day
        self.m = month
        self.y = year
    def leap_year(self):
        return self.y%400==0 or (self.y%4==0 and self.y%100!=0)
    def day_in_month(self):
        m31 = [1,3,5,7,8,10,12]
        m30 = [4,6,9,11]
        if self.m in m30:
            return 30
        if self.m in m31:
            return 31
        if self.m==2:
            if self.leap_year():
                return 29
            else:
                return 28
        return -1
    def check_valid(self):
        numDay = self.day_in_month()
        if numDay == -1:
            return False
        if 1<=self.d<=numDay:
            return True
        return False
    def next_date(self):
        if self.check_valid():
            numDay = self.day_in_month()
            if self.d<numDay:
                print(self.d+1,"/",self.m,"/",self.y,sep="")
            else:
                if self.m+1<13:
                    print("1/",self.m+1,"/",self.y,sep="")
                else:
                    print("1/1/",self.y+1,sep="")
        else:
            print("INVALID")

inp = [int(x) for x in input().split(sep = "/")]
date = Date(inp[0], inp[1], inp[2])
date.next_date()
```

```python
#W8A3
from math import sqrt

class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    def is_origin(self):
        return self.x==0 and self.y==0
    def is_on_x_axis(self):
        return self.y==0
    def is_on_y_axis(self):
        return self.x ==0
    def dist(self, other):
        delta_x = self.x - other.x
        delta_y = self.y - other.y
        return sqrt(delta_x*delta_x + delta_y*delta_y)

xm, ym = map(int, input().split())
M = Point(xm, ym)
origin = Point(0,0)
if M.is_origin():
    print("M la goc toa do")
elif M.is_on_x_axis():
    print("M nam tren truc hoanh")
elif M.is_on_y_axis:
    print("M nam tren truc tung")
else:
    print("M khong la goc toa do, khong nam tren truc tung hoac truc hoanh")
print(f"{M.dist(origin):.2f}")
```

```python
#W8A4
class Calculator:
    def __init__(self, num1, num2):
        self.num1 = num1
        self.num2 = num2
    def add(self):
        return self.num1 + self.num2
    def subtract(self):
        return self.num1 - self.num2
    def product(self):
        return self.num1 * self.num2
    def divide(self):
        return self.num1 / self.num2
    def power(self):
        n = self.num1
        k = self.num2
        ans = 1
        while k:
            if k&1:
                ans*=n
            n*=n
            k>>=1
        return ans
    def mod(self):
        return self.num1%self.num2
    def set_number(self, num1, num2):
        self.num1 = num1
        self.num2 = num2
print("Nhap 2 so can tinh: ", end="")
num1, num2 = map(int, input().split())
calc = Calculator(num1, num2)
while True:
    print("Chon phep tinh cua ban: ")
    print("1, Cong")
    print("2, Tru")
    print("3, Nhan")
    print("4, Chia")
    print("5, Luy thua")
    print("6, Chia lay du")
    print("7, Dat 2 so moi")
    opt = int(input())
    if opt==1:
        print(calc.add())
    elif opt==2:
        print(calc.subtract())
    elif opt==3:
        print(calc.product())
    elif opt==4:
        print(calc.divide())
    elif opt==5:
        print(calc.power())
    elif opt == 6:
        print(calc.mod())
    elif opt == 7:
        print("Nhap 2 so moi: ", end="")
        num1, num2 = map(int, input().split())
        calc.set_number(num1, num2)
    print("Ban co muon thoat?")
    opt = input()
    if opt == "yes" or opt == "YES" or opt == "Yes":
        break
```

```python
#W8A5
class ShoppingCart:
    def __init__(self):
        self.cart = []
    def add_to_cart(self, product):
        self.cart.append(product)
    def remove_from_cart(self, product):
        for idx, i in enumerate(self.cart):
            if i['name'] == product:
                self.cart.pop(idx)
    def check_empty(self):
        return len(self.cart)==0
    def total(self):
        ans = 0
        for i in self.cart:
            ans+=i['price']
        return ans
    def display_cart(self):
        for i in self.cart:
            print(i['name'], i['price'])
    def empty_cart(self):
        self.cart = []

cart = ShoppingCart()
while True:
    print("Chon hanh dong cua ban:")
    print("1, Them san pham vao gio hang")
    print("2, Xoa san pham khoi gio hang")
    print("3, Kiem tra gio hang trong")
    print("4, Tinh tong tien cua gio hang")
    print("5, Hien thi cac san pham trong gio")
    print("6, Xoa toan bo gio hang")
    opt = int(input())
    if opt==1:
        print("Nhap ten san pham: ", end="")
        product_name = input()
        print("Nhap gia san pham: ", end="")
        product_price = float(input())
        cart.add_to_cart({'name':product_name, 'price':product_price})
    elif opt==2:
        print("Nhap ten san pham: ", end="")
        product_name = input()
        cart.remove_from_cart(product_name)
    elif opt==3:
        if cart.check_empty():
            print("Gio hang trong")
        else:
            print("Gio hang khong trong")
    elif opt==4:
        print("Tong tien:", cart.total())
    elif opt==6:
        cart.display_cart()
    elif opt==7:
        cart.empty_cart()
    print("Ban co muon thoat?")
    opt = input()
    if opt == "yes" or opt == "YES" or opt == "Yes":
        break
```

```python
#W8A6
class Rectangle:
    def __init__(self, w, h):
        self.w = w
        self.h = h
    def area(self):
        return self.w*self.h
    def perimeter(self):
        return (self.h+self.w)*2
    def scale(self, k):
        self.h*=k
        self.w*=k

w,h,k = map(int, input().split())
rec = Rectangle(w,h)
rec.scale(k)
print(rec.area(), rec.perimeter())
```

```python
#W8A7
class Fraction:
    def __init__(self, num, den):
        self.num = num
        self.den = den
        self.reduce()
    def __str__(self):
        return f"{int(self.num)}/{int(self.den)}"
    def gcd(self, a, b):
        while b:
            a,b = b, a%b
        return a
    def reduce(self):
        r = self.gcd(self.num, self.den)
        self.num/=r
        self.den/=r
    def add(self, other):
        self.num = self.num*other.den + self.den*other.num
        self.den*=other.den
        self.reduce()
    def sub(self, other):
        self.num = self.num*other.den - self.den*other.num
        self.den*=other.den
        self.reduce()
    def mul(self, other):
        self.num*=other.num
        self.den*=other.den
        self.reduce()
    def div(self, other):
        self.num*=other.den
        self.den*=other.num
        self.reduce()

a,b,op,c,d = input().split()
a = int(a)
b = int(b)
c = int(c)
d = int(d)
frac1 = Fraction(a, b)
frac2 = Fraction(c, d)
if op=='+':
    frac1.add(frac2)
elif op=='-':
    frac1.sub(frac2)
elif op=='*':
    frac1.mul(frac2)
else:
    frac1.div(frac2)
print(frac1)
```

```python
#W8A8
class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner
        self.balance = balance
    def __str__(self):
        return self.owner+f": {self.balance}"
    def deposit(self, x):
        self.balance += x
    def withdraw(self, x):
        if x<=self.balance:
            self.balance -= x
    def get_balance(self):
        return self.balance

owner, initial_balance = input().split()
initial_balance = int(initial_balance)
bankacc = BankAccount(owner, initial_balance)
n = int(input())
for i in range(n):
    cmd, val = input().split()
    val = int(val)
    if cmd == 'DEPOSIT':
        bankacc.deposit(val)
    else:
        bankacc.withdraw(val)

print(bankacc.get_balance())
```

```python
#W8A9
class Student:
    def __init__(self, name):
        self.name = name
        self.ss = {}
    def add_score(self, subject, score):
        self.ss[subject] = score
    def avg(self):
        sum = 0
        sz = 0
        for i in self.ss.values():
            sum+=i
            sz+=1
        return sum/sz
    def rank(self):
        score = self.avg()
        if score>=8:
            return 'Excellent'
        if 6.5<=score<8:
            return 'Good'
        if 5<=score<6.5:
            return 'Average'
        return 'Poor'

name = input()
student = Student(name)
n = int(input())
for i in range(n):
    subject, score = input().split()
    score = float(score)
    student.add_score(subject, score)

print(student.name, student.avg(), student.rank())
```

```python
#W8A10
class Book:
    def __init__(self, title, author, year):
        self.title = title
        self.author = author
        self.year = year

class Library:
    def __init__(self):
        self.lib = []
    def add(self, book):
        self.lib.append(book)
    def count_by_author(self, name):
        ans = 0
        for book in self.lib:
            if book.author == name:
                ans+=1
        return ans
    def find_by_year(self, y):
        ans  = 0
        for book in self.lib:
            if book.year == y:
                ans+=1
        return ans

library = Library()
n= int(input())
for i in range(n):
    q, data = input().split()
    if q=='ADD':
        title, author, year = data.split(';')
        library.lib.append(Book(title, author, year))
    elif q == 'COUNT':
        print(library.count_by_author(data))
    else:
        print(library.find_by_year(data))
```

```python
#W8A11
import math

class Point:
    def __init__(self, x, y, z):
        self.x = x
        self.y = y
        self.z = z

class Vector:
    def __init__(self, point1, point2):
        self.x = point2.x - point1.x
        self.y = point2.y - point1.y
        self.z = point2.z - point1.z
    def length(self):
        return math.sqrt(self.x*self.x + self.y*self.y + self.z*self.z)
    def angle_with_vector(self, other):
        dotprod = self.x*other.x + self.y*other.y + self.z*other.z
        len1 = self.length()
        len2 = other.length()
        cos = dotprod / (len1*len2)
        radian = math.acos(cos)
        return radian*(180/math.pi)
    def cross_product(self, other):
        x = self.y*other.z - self.z*other.y
        y = self.z*other.x - self.x*other.z
        z = self.x*other.y - self.y*other.x
        point1 = Point(0,0,0)
        point2 = Point(x,y,z)
        return Vector(point1, point2)
class Surface:
    def __init__(self, point1, point2, point3):
        v1 = Vector(point1, point2)
        v2 = Vector(point1, point3)
        self.normal_vector = v1.cross_product(v2)
        self.point = point1
    def angle_with_surface(self, other):
        angle = self.normal_vector.angle_with_vector(other.normal_vector)
        angle = abs(angle)
        return f"{min(angle, (180-angle)):.2f}"

point = []

for i in range(4):
    x,y,z = map(int, input().split())
    point.append(Point(x,y,z))

surface1 = Surface(point[0], point[1], point[2])
surface2 = Surface(point[1], point[2], point[3])

print(surface1.angle_with_surface(surface2))
```

```python
#W8A12
class ComplexNumber:
    def __init__(self, realPart, imaginaryPart):
        self.realPart = realPart
        self.imaginaryPart = imaginaryPart
    def multiply(self, other):
        realPart = self.realPart*other.realPart - self.imaginaryPart*other.imaginaryPart
        imaginaryPart = self.realPart*other.imaginaryPart + self.imaginaryPart*other.realPart
        return ComplexNumber(realPart, imaginaryPart)
    def display(self):
        print(self.realPart, self.imaginaryPart)

a, b = map(float, input().split())
cn1 = ComplexNumber(a, b)
c, d = map(float, input().split())
cn2 = ComplexNumber(c, d)
prod = cn1.multiply(cn2)
prod.display()
```

```python
#W8A13
import math

class Triangle:
    def __init__(self, edge1, edge2, edge3):
        self.edge1 = edge1
        self.edge2 = edge2
        self.edge3 = edge3
    def getArea(self):
        a = self.edge1
        b= self.edge2
        c = self.edge3
        if a+b<=c or b+c<=a or a+c<=b:
            return -1
        p = (a+b+c)/2
        return f"{math.sqrt(p*(p-a)*(p-b)*(p-c)):.2f}"

a, b, c = map(int, input().split())
triangle = Triangle(a,b,c)
area = triangle.getArea()
if area==-1:
    print('invalid')
else:
    print(area)
```
